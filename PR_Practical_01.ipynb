```python
# Step 1: Set up the environment
# Google Colab already has most libraries like numpy and matplotlib installed.

# Step 2: Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Step 3: Generate sample data
# Let's create a dataset with a linear relationship
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Step 4: Define the model and loss function
# Linear regression model: y = theta0 + theta1 * X
def predict(X, theta):
    return X.dot(theta)

# Mean Squared Error Loss Function
def compute_loss(y, y_pred):
    m = len(y)
    return (1 / (2 * m)) * np.sum((y_pred - y) ** 2)

# Step 5: Implement gradient descent
def gradient_descent(X, y, theta, learning_rate=0.01, n_iterations=1000):
    m = len(y)
    X_b = np.c_[np.ones((m, 1)), X]  # Add bias term (X0 = 1)
    loss_history = []

    for iteration in range(n_iterations):
        gradients = 1 / m * X_b.T.dot(X_b.dot(theta) - y)
        theta -= learning_rate * gradients
        y_pred = X_b.dot(theta)
        loss = compute_loss(y, y_pred)
        loss_history.append(loss)

    return theta, loss_history

# Initial parameters
theta = np.random.randn(2, 1)  # Random initialization

# Step 6: Train the model
theta, loss_history = gradient_descent(X, y, theta)

# Step 7: Visualize the results
plt.plot(range(len(loss_history)), loss_history)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss reduction over time')
plt.show()

# Plot the data and the model
plt.scatter(X, y, label='Data')
plt.plot(X, predict(np.c_[np.ones((X.shape[0], 1)), X], theta), color='red', label='Linear model')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression Fit')
plt.show()
```
